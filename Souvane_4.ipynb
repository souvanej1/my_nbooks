```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mortality.csv')
```

```python
plt.figure(figsize=(1.8, 1.8))
avg_death_rates = df.groupby(['Year', 'Age Group'])['Death Rate'].mean().unstack()
avg_death_rates.plot(kind='bar', stacked=True)
plt.title('Avg Death Rates by Age Group (1900, 1950, 2000)')
plt.xlabel('Year')
plt.ylabel('Deaths per 100,000')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))
years = [1900, 1950, 2000]
for i, ax in enumerate(axes.flatten()):
    year_data = df[df['Year'] == years[i]]
    year_data.groupby('Age Group')['Death Rate'].mean().plot(kind='bar', ax=ax)
    ax.set_title(f'Death Rates for {years[i]}')
    ax.set_xlabel('Age Group')
    ax.set_ylabel('Deaths per 100,000')
    ax.grid(axis= 'y')
```

```python
plt.figure(figsize=(10, 6))
agegroup_15_19 = df[df['Age Group'] == '15-19']
plt.plot(age_group_15_19['Year'], agegroup_15_19['Death Rate'])
plt.title('Death Rates for Age Group 15-19 Over the Years')
plt.xlabel('Year')
plt.ylabel('Deaths per 100,000')
plt.grid(axis= 'y')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
death_rates_1950_2000 = df[df['Year'].between(1950, 2000)]
death_rates_pivot = death_rates_1950_2000.pivot_table(index='Age Group', columns='Year', values='Death Rate')
death_rates_pivot.plot(kind='line')
plt.title('Death Rates by Age Group (1950-2000)')
plt.xlabel('Age Group')
plt.ylabel('Deaths per 100,000')
plt.grid(axis= 'y')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(death_rates_1950_2000['Year'], death_rates_1950_2000['Death Rate'], s=100, alpha=0.7)
plt.title('Scatter Plot of Death Rates (1950-2000)')
plt.xlabel('Year')
plt.ylabel('Deaths per 100,000')\
plt.grid(axis= 'y')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))
years = [1900, 1925, 1950, 1975, 2000]
for i, ax in enumerate(axes.flatten()):
    year_data = df[df['Year'] == years[i]]
    year_data.groupby('Age Group')['Death Rate'].mean().plot(kind='bar', ax=ax)
    ax.set_title(f'Death Rates for {years[i]}')
    ax.set_xlabel('Age Group')
    ax.set_ylabel('Deaths per 100,000')
    ax.grid(axis='y')
```
